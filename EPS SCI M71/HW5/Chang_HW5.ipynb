## Problem 1 (Working with Hurricane and Storm Data)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Problem 1.1

```python
yr, storm, hurr, maj_hurr = np.loadtxt("AtlanticStorms.txt", unpack=True, skiprows = 2) #unpacking data into four arrays: years, number of tropical storms, number of hurricanes, and number of major hurricanes
```

### Problem 1.2

```python
print(yr) #printing year array
```

```python
print(storm) #printing storm array
```

```python
print(hurr)  #printing hurricane array
```

```python
print(maj_hurr) #printing major hurricane array
```

### Problem 1.3

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, figsize=(11,11), sharex=True, sharey=True) #setting up the figure

#first subplot
ax1.plot(yr, storm, "ro", markerfacecolor = "None", label = "Storms") #storms vs year, open red circles
ax1.plot(yr, hurr, "b^", markerfacecolor = "None", label = "Hurricanes") #hurricanes vs year, open blue triangles
ax1.plot(yr, maj_hurr, "gD", markerfacecolor = "None", label = "Major Hurricanes") #major hurricanes vs year, open green diamonds

ax1.legend(loc="upper left") #legend in upper left
ax1.grid()

ax1.set_xlabel("Time (Years)") #x-axis: time
ax1.set_ylabel("Number") #y-axis: number

#second subplot
before_1970_BA = (yr < 1970)
before_1970 = storm[before_1970_BA] #storms before 1970
before_1970_yr = yr[before_1970_BA] #years before 1970

after_1970_BA = (yr >= 1970)
after_1970 = storm[after_1970_BA] #storms during and after 1970
after_1970_yr = yr[after_1970_BA] #1970 and after

ax2.plot(before_1970_yr, before_1970, "ro", markerfacecolor = "None", label = "Before 1970") #storms before 1970, open red circles
ax2.plot(after_1970_yr, after_1970, "ro", label = "During & After 1970") #storms during and after 1970, filled red circles

ax2.legend(loc="upper left")
ax2.grid()

ax2.set_xlabel("Time (Years)")
ax2.set_ylabel("Number of Storms")

#third subplot
before_1970 = hurr[before_1970_BA] #hurricanes before 1970

after_1970 = hurr[after_1970_BA] #hurricanes during and after 1970

ax3.plot(before_1970_yr, before_1970, "b^", markerfacecolor = "None", label = "Before 1970") #hurricanes before 1970, open blue triangles
ax3.plot(after_1970_yr, after_1970, "b^", label = "During & After 1970") #storms during and after 1970, filled blue triangles

ax3.legend(loc="upper left")
ax3.grid()

ax3.set_xlabel("Time (Years)")
ax3.set_ylabel("Number of Hurricanes")

#fourth subplot
before_1970 = maj_hurr[before_1970_BA] #major hurricanes before 1970

after_1970 = maj_hurr[after_1970_BA] #major hurricanes during and after 1970

ax4.plot(before_1970_yr, before_1970, "gD", markerfacecolor = "None", label = "Before 1970") #major hurricanes before 1970, open green diamonds
ax4.plot(after_1970_yr, after_1970, "gD", label = "During & After 1970") #storms during and after 1970, filled green diamonds

ax4.legend(loc="upper left")
ax4.grid()

ax4.set_xlabel("Time (Years)")
ax4.set_ylabel("Number of Major Hurricanes")


#subplots title
plt.suptitle("Cyclone Data; M. Chang, 11/10/20", y = 0.92)

plt.savefig("Chang_HurrStormSubplots.png", dpi=200)
```

## Problem 2

### Problem 2.1

```python
name, age, depth = np.loadtxt("MOR_data.txt", unpack = True, dtype = str) #reading in MOR_data into columns as str
```

### Problem 2.2

```python
#converting age and depths to floats
age = np.asarray(age, dtype=np.float)
depth = np.asarray(depth, dtype=np.float)
```

```python
print(name) #printing name
            #variable type is string
```

```python
plt.figure(2, figsize = (7,5)) #figure window

#Mid-Atlantic Ridge
MAR_BA = (name == "MAR")
MAR = depth[MAR_BA] #depths corresponding to MAR
MAR_year = age[MAR_BA] #ages corresponding to MAR
plt.plot(MAR_year, MAR, "bs", label = "Mid-Atlantic Ridge") #depth vs yr, blue squares

#East Pacific Rise
EPR_BA = (name == "EPR")
EPR = depth[EPR_BA] #depths corresponding to EPR
EPR_year = age[EPR_BA] #ages corresponding to EPR
plt.plot(EPR_year, EPR, "ro", label = "East Pacific Rise") #depth vs yr, red circles

#Central Indian Ridge
CIR_BA = (name == "CIR") 
CIR = depth[CIR_BA] #depths corresponding to CIR
CIR_year = age[CIR_BA] #ages corresponding to CIR
plt.plot(CIR_year, CIR, "m*", label = "Central Indian Ridge") #depth vs yr, magenta stars

plt.gca().invert_yaxis() #reversing y-axis, so largest depths at bottom
plt.xlim(np.min(age) - 5, np.max(age) + 5) #limiting x-axis

after_75_BA = (age > 75)
after_75 = depth[after_75_BA]
plt.hlines(xmin = 75, xmax = 180, y = np.mean(after_75), color = "green", ls = "--", label = "Average Depth after t = 75 mya") #from 75-180 mya, the average of the MOR depths

plt.legend(loc="upper right") #legend in upper right corner
plt.grid(linewidth = 0.5) #width of grid lines is 0.5

plt.title("Age vs Depth of Mid-Ocean Ridge, M. Chang, 11/10/20")
plt.xlabel("Age (mya)")
plt.ylabel("Depth (m)")

plt.savefig("Chang_MOR_ColorCoded.png", dpi = 200)
```

## Problem 3 (Balloon Data)

### Problem 3.1

```python
t, alt = np.loadtxt("BalloonData.txt", unpack=True) #reading in data to time and altitude arrays
```

### Problem 3.2

```python
print(len(t)) #length of time array
print(len(alt)) #length of altitude array
```

### Problem 3.3

```python
plt.plot(t, alt) #raw figure
```

### Problem 3.4

```python
actual_data = (alt > 475) #cutting out the noise and time before launch
Alt = alt[actual_data]
tsec = t[actual_data]
```

### Problem 3.5

```python
dAlt = np.diff(Alt) #difference in altitude
dtsec = np.diff(tsec) #difference in time
vel = dAlt/dtsec #velocity
dtsec_mean = np.mean(dtsec) #mean of dtsec

dvel = np.diff(vel) #difference in velocity
acc = dvel/(dtsec[0:len(dtsec)-1]) #acceleration
```

### Problem 3.6

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(10,10))

#first subplot
n = ax1
n.plot(tsec, Alt, "bo", markerfacecolor = "None") #altitude v time, open blue circles
n.set_xlabel("Time (s)")
n.set_ylabel("Altitude (m)")
n.grid()

#second subplot
n = ax2
n.plot(tsec[0:len(tsec)-1], vel, "rs", markerfacecolor = "None") #velocity v time, open blue circles
n.set_xlabel("Time (s)")
n.set_ylabel("Velocity (m/s)")
n.grid()

#third subplot
n = ax3
n.plot(tsec[0:len(tsec)-2], acc, color = "black", marker = "^", ls="None", markerfacecolor = "None") #acceleration v time, open blue circles
n.set_xlabel("Time (s)")
n.set_ylabel("Acceleration (m/s^2)")
n.grid()

plt.suptitle("Balloon Data; M. Chang, 11-10-20, Mean Sample Period = {0:0.2f} s".format(dtsec_mean), y=0.92)

plt.savefig("Chang_BalloonMagic.png", dpi=200)
```

### Problem 3.7

Does the mean balloon rise velocity (= 6 m/s) multiplied by the rise time (= 575 s) approximately match (max(Altitude) - min(Altitude))? Does this make sense? </br>
<font color = 'red'>Yes, max(Altitude) - min(Altitude) = 3577 m while 6*575 is 3450. This makes sense since how much the balloon traveled should equal its speed multiplied by the time elapsed</font>

Why is the accleration nearly zero everywhere, except when the cord is cut (-4 m/s2)? </br>
<font color = 'red'>Because the balloon is traveling at constant speed. When the cord is cut, the balloon changes direction (and falls due to gravity), which explains why there is an acceleration.</font>

Can we ascertain why the rise velocity is so close to the descent velocity? </br>
<font color = 'red'>Yes as shown by the graphs, the distance ascended is equal to the distance descended over equal periods of time, so the rise velocity is bound to be close to the descent velocity.</font>
